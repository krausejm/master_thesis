```python
def makebin(phi,sigma,pol_p45,pol_m45,n1,n2):    
    #p45 setting
    f_np45=1-sigma*pol_p45*np.cos(2*(+45-phi)*np.pi/180.)
    #m45 setting
    f_nm45=1-sigma*pol_m45*np.cos(2*(-45-phi)*np.pi/180.)
    
    #draw random from distribution
    np45=random.choices(phi,f_np45,k=np.random.poisson(n1))
    nm45=random.choices(phi,f_nm45,k=np.random.poisson(n2))
    
    setting=[+45 for i in range(len(np45))]+[-45 for i in range(len(nm45))]
    
    pol=[pol_p45 for i in range(len(np45))]+[pol_m45 for i in range(len(nm45))]
    
    phi=np45+nm45
    
    df=pd.DataFrame({'pol': pol,
                    'setting':setting,
                    'phi':phi})
    return df
    
```

```python
#create grid in phi
phi=np.linspace(-180,180,100000)
```

```python
for i in range(10000):
    if(i%100==0):
        print(i)
    toybin=makebin(phi,0.5,0.3,0.25,10000,8000)
    toybin.to_csv(f'./py_toybins/toybin{i:04d}.txt',index=False,header=False)
```
